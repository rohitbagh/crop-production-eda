```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\USER\OneDrive\Desktop\DATA CROP\Crop Production data - Crop Production data.csv')
```

```python

df.info()
```

```python
df.isnull().sum()
```

```python
df_clean = df.dropna(subset=['Production']).copy()
```

```python
print("Cleaned shape:", df_clean.shape)
```

```python
stats = df_clean.describe()
print(" Descriptive Statistics\n", stats)
```

```python
plt.figure(figsize=(8, 4))
top_crops = df_clean['Crop'].value_counts().head(10)
sns.barplot(x=top_crops.values, y=top_crops.index)
plt.title('Top 10 Most Frequently Grown Crops')
plt.xlabel('Count')
plt.ylabel('Crop')
plt.show()

```

```python
plt.figure(figsize=(8, 4))
state_production = df_clean.groupby('State_Name')['Production'].sum().sort_values(ascending=False).head(10).reset_index()
sns.barplot(data=state_production, x='Production', y='State_Name')
plt.title('Top 10 States by Total Production')
plt.xlabel('Total Production')
plt.ylabel('State')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
yearly_production = df_clean.groupby('Crop_Year')['Production'].sum().reset_index()
sns.lineplot(data=yearly_production, x='Crop_Year', y='Production', marker='o', color='green')
plt.title('Total Crop Production Over Years')
plt.xlabel('Year')
plt.ylabel('Total Production')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df_clean, x='Season', order=df_clean['Season'].value_counts().index)
plt.title('Number of Crop Records per Season')
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```
